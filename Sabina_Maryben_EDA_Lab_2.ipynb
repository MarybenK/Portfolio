# Lab Activity One: Guided Data Cleaning

##### It is generally a good idea to make a copy of the master dataset for cleaning so you can always go back if ever needed.

```python
df_clean = df1.copy()
```

##### [A] Drop the `'Unnamed: 0'` column

```python
# Drop the 'Unnamed: 0' column
df_clean = df_clean.drop('Unnamed: 0', axis=1)
```

##### [A] Fix the data in the `'Gender'` column.
> Hint: make it so there are 2 unique entries, `'male'` and `'female'`.

```python
# Fix the 'Gender' column - standardize to 'male' and 'female'
df_clean['Gender'] = df_clean['Gender'].str.strip().str.lower()  # removes leading/trailing spaces and convert all to lowercase
df_clean['Gender'] = df_clean['Gender'].replace({
     'male': 'male', 'Male ': 'male',
     'fmeale': 'female', 'female ': 'female'
})  # Fix typo and corrects both typos and whitespace issues.
```

Check the 'Gender' column again

```python
df_clean['Gender'].unique()
```

##### [A] Fix the data in the `'Mode'` column.
> Hint: Check for lower and upper case entries

```python
# Fix the 'Mode' column - standardize case and handle missing values
df_clean['Mode'] = df_clean['Mode'].str.strip().str.upper()  # Convert all to uppercase
df_clean['Mode'] = df_clean['Mode'].replace('_', pd.NA)  # Replace underscores with NaN
print("Cleaned Mode unique values):", df_clean['Mode'].unique())
```

### Setting The Correct Data Type

##### [A] Change the `'H'` column to numeric data `type(float64)` instead of `'object'`.

```python
# Convert 'H' column to numeric, handling non-numeric values
df_clean['H'] = pd.to_numeric(df_clean['H'], errors='coerce')
print(df_clean['H'].head())
```

*Check* the 'H' column again

```python
df_clean['H'].dtype
```

##### [A] Convert `'EDate'` column to `datetime` type.

```python
# Convert 'EDate' to datetime
df_clean['EDate'] = pd.to_datetime(df_clean['EDate'], format='%m/%d/%Y')
```

```python
df_clean.dtypes
```

##### Using the `gender_type` key we set the `'Gender'` column as categorical data.

```python
gender_type = pd.CategoricalDtype(categories=["female", "male"])

df_clean["Gender"] = df_clean["Gender"].astype(gender_type)
```

```python
df_clean.dtypes
```

##### [A] Replicate how we changed the gender column to categorical but this time for the `'Mode'` column.

```python
# Convert 'Mode' to categorical
mode_type = pd.CategoricalDtype(categories=["D", "E", "F", "G", "H", "A"])
df_clean["Mode"] = df_clean["Mode"].astype(mode_type)
df_clean.dtypes
```

##### [A] Again, change the `'Result'` column to categorical replicating above the example above.

```python
# Convert 'Result' to categorical
result_type = pd.CategoricalDtype(categories=["Negative", "Positive"])
df_clean["Result"] = df_clean["Result"].replace("_", pd.NA)  # Clean invalid values first
df_clean["Result"] = df_clean["Result"].astype(result_type)
df_clean.dtypes
```

### Duplicate Entries
- Duplicate entities in a dataset is not good to have as it  can use overfit and is redundant information

##### [A] Check for duplicate entries and delete them

```python
duplicate_rows = df_clean.duplicated()
```

```python
display(duplicate_rows)
```

```python
df_clean = df_clean[~duplicate_rows]
```

##### [A] Once you delete a row in your dataset the index of that row is also deleted. Reset the index of the dataset so it in proper order.
> Hint: use the `dataframe.reset_index` function and set the drop parameter to `True`.

```python
# Reset index after deleting rows
df_clean = df_clean.reset_index(drop=True)
```

In this activity, you will need to clean the dataset yourself using the examples from the activity above. The `Lab2_df2.csv` is loaded and displayed for you.

```python
#Read Lab2_df2.csv file

df2 = pd.read_csv('Lab2_df2.csv')
df2.head(10)
```

```python
print("Dataset Info:")
print(df2.info())
print(f"\nDataset Shape: {df2.shape}")
```

##### [A] Figure out the issues with this dataset and apply the cleaning methods

```python
# [A] Figure out the issues with this dataset and apply cleaning methods

# Make a copy for cleaning
df2_clean = df2.copy()

# Overview of the data types
print(df2_clean.dtypes)
```

```python
# Issue 1: Drop the unnecessary 'Unnamed: 0' column
df2_clean = df2_clean.drop('Unnamed: 0', axis=1)
```

```python
# Issue 2: Convert 'Inspection Date' to datetime
df2_clean['Inspection Date'] = pd.to_datetime(df2_clean['Inspection Date'], format='%m/%d/%Y')
print(df2_clean['Inspection Date'].head())
```

```python
# Issue 3: Fix M1 column - replace '_' with NaN and convert to numeric
df2_clean['M1'] = df2_clean['M1'].replace('_', pd.NA)
df2_clean['M1'] = pd.to_numeric(df2_clean['M1'], errors='coerce')
print(df2_clean['M1'].head())
```

```python
# Issue 4: Convert 'Site Name' to categorical
site_type = pd.CategoricalDtype(categories=["Clinic1", "Clinic2"])
df2_clean['Site Name'] = df2_clean['Site Name'].astype(site_type)
print(df2_clean['Site Name'].head())
```

```python
# Issue 5: Check for duplicates and remove them
df2_clean = df2_clean.drop_duplicates()
print(df2_clean.shape)
```

```python
# Issue 6: Reset index
df2_clean = df2_clean.reset_index(drop=True)
print(df2_clean.head())
```

```python
# Show the clean data
print("Cleaned Dataset:")
df2_clean.head(10)
```

```python
# Check the data types and shape
print("Data Types after cleaning:")
print(df2_clean.dtypes)

print(f"\nShape of cleaned dataset: {df2_clean.shape}")
```

### Issues Found and Fixed:
- Unnecessary column: 'Unnamed: 0' - dropped

- Date format: 'Inspection Date' - converted from string to datetime

- Missing values in M1: '_' values - replaced with NaN

- Categorical data: 'Site Name' - converted to categorical type

- Duplicates: Checked and removed any duplicate rows

- Index: Reset to proper order
